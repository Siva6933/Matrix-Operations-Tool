```python
import numpy as np

def input_matrix(name="Matrix"):
    rows = int(input(f"Enter number of rows for {name}: "))
    cols = int(input(f"Enter number of columns for {name}: "))
    print(f"Enter elements row by row for {name}:")
    matrix = []
    for i in range(rows):
        row = list(map(float, input(f"Row {i+1}: ").split()))
        if len(row) != cols:
            print("Error: Number of elements must match columns.")
            return input_matrix(name)
        matrix.append(row)
    return np.array(matrix)

def display_matrix(matrix, name="Result"):
    print(f"\n{name}:\n{matrix}\n")

def main():
    print("=== Matrix Operations Tool ===")
    print("Operations available:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Transpose")
    print("5. Determinant")
    print("6. Exit")

    while True:
        choice = input("\nEnter your choice (1-6): ")

        if choice == '1':
            A = input_matrix("Matrix A")
            B = input_matrix("Matrix B")
            if A.shape == B.shape:
                display_matrix(A + B, "A + B")
            else:
                print("Error: Matrices must be of same dimension for addition.")

        elif choice == '2':
            A = input_matrix("Matrix A")
            B = input_matrix("Matrix B")
            if A.shape == B.shape:
                display_matrix(A - B, "A - B")
            else:
                print("Error: Matrices must be of same dimension for subtraction.")

        elif choice == '3':
            A = input_matrix("Matrix A")
            B = input_matrix("Matrix B")
            if A.shape[1] == B.shape[0]:
                display_matrix(np.matmul(A, B), "A x B")
            else:
                print("Error: Columns of A must match rows of B for multiplication.")

        elif choice == '4':
            A = input_matrix("Matrix")
            display_matrix(A.T, "Transpose")

        elif choice == '5':
            A = input_matrix("Matrix")
            if A.shape[0] == A.shape[1]:
                display_matrix(np.linalg.det(A), "Determinant")
            else:
                print("Error: Determinant is only defined for square matrices.")

        elif choice == '6':
            print("Exiting Matrix Operations Tool. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
